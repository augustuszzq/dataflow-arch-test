```python
import csv
import re

def convert_table_to_csv(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as fin:
        lines = fin.readlines()

    rows = []
    for line in lines:
        line = line.strip()
        if line.startswith('|') and line.endswith('|'):
            if re.match(r'^\|[-+\s]+$', line):
                continue
            cells = [cell.strip() for cell in line.split('|')[1:-1]]
            if any(cells):
                rows.append(cells)

    with open(output_file, 'w', encoding='utf-8', newline='') as fout:
        writer = csv.writer(fout)
        for row in rows:
            writer.writerow(row)
convert_table_to_csv('compile_report_22.txt', 'compile_report_22.csv')
#convert_table_to_csv('compile_report_20.txt', 'compile_report_20.csv')
#convert_table_to_csv('compile_report_19.txt', 'compile_report_19.csv')
#convert_table_to_csv('compile_report_18.txt', 'compile_report_18.csv')
#convert_table_to_csv('compile_report_17.txt', 'compile_report_17.csv')
#convert_table_to_csv('compile_report_16.txt', 'compile_report_15.csv')
#convert_table_to_csv('compile_report_15.txt', 'compile_report_16.csv')
#convert_table_to_csv('compile_report_14.txt', 'compile_report_14.csv')
#convert_table_to_csv('compile_report_13.txt', 'compile_report_13.csv')
#convert_table_to_csv('compile_report_11.txt', 'compile_report_11.csv')
#convert_table_to_csv('compile_report_7.txt', 'compile_report_7.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
chapters = [7,11,13,14, 15, 16, 17, 18, 19, 20,22]
tp = [5887,5897,5897,5237,4560,3387,2026,1357,930,636,287]
def load_data(metric_type):

    data = {}
    for ch in chapters:
        filename = f'filtered_compile_report_{ch}_{metric_type}.csv'
        df = pd.read_csv(filename)
        df['Samples/s'] = pd.to_numeric(df['Samples/s'], errors='coerce')
        df['Active PEs'] = pd.to_numeric(df['Active PEs'], errors='coerce')
        df['Compute Utilization'] = df['Compute Utilization'].str.rstrip('%')
        df['Compute Utilization'] = pd.to_numeric(df['Compute Utilization'], errors='coerce')
        data[ch] = df
    return data

data_fc = load_data('fc')
data_attention = load_data('attention')

def collect_metric(data, column):

    return [data[ch][column].dropna() for ch in chapters]

samples_fc = collect_metric(data_fc, 'Samples/s')
active_pe_fc = collect_metric(data_fc, 'Active PEs')
util_fc = collect_metric(data_fc, 'Compute Utilization')

samples_attention = collect_metric(data_attention, 'Samples/s')
active_pe_attention = collect_metric(data_attention, 'Active PEs')
util_attention = collect_metric(data_attention, 'Compute Utilization')


plt.figure(figsize=(10,6))
plt.boxplot(samples_fc, labels=chapters)
plt.title('FC Samples/s Across Chapters')
plt.scatter(range(1, len(tp) + 1), tp, color='red', marker='o', s=100, label="TP Data")
plt.xlabel('Chapter')
plt.ylabel('Samples/s')
plt.tight_layout()
plt.savefig('fc_samples_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(active_pe_fc, labels=chapters)
plt.title('FC Active PEs Across Chapters')
plt.xlabel('Chapter')
plt.ylabel('Active PEs')
plt.tight_layout()
plt.savefig('fc_active_pe_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(util_fc, labels=chapters)
plt.title('FC Compute Utilization (%) Across Chapters')
plt.xlabel('Chapter')
plt.ylabel('Compute Utilization (%)')
plt.tight_layout()
plt.savefig('fc_util_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(samples_attention, labels=chapters)
plt.title('Attention Samples/s Across Chapters')

plt.scatter(range(1, len(tp) + 1), tp, color='red', marker='o', s=100, label="TP Data")
plt.xlabel('Chapter')
plt.ylabel('Samples/s')
plt.tight_layout()
plt.savefig('attention_samples_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(active_pe_attention, labels=chapters)
plt.title('Attention Active PEs Across Chapters')
plt.xlabel('Chapter')
plt.ylabel('Active PEs')
plt.tight_layout()
plt.savefig('attention_active_pe_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(util_attention, labels=chapters)
plt.title('Attention Compute Utilization (%) Across Chapters')
plt.xlabel('Chapter')
plt.ylabel('Compute Utilization (%)')
plt.tight_layout()
plt.savefig('attention_util_boxplot.png')
plt.show()
```

```python
import numpy as np
import pandas as pd
import glob

# 
fc_files = glob.glob("filtered_compile_report_*_fc.csv")
attention_files = glob.glob("filtered_compile_report_*_attention.csv")

def sum_active_pes(files):
    results = {}
    for file in files:
        df = pd.read_csv(file)
        if 'Active PEs' in df.columns:
            results[file] = df['Active PEs'].sum()
    return results

#  Active PEs 
fc_sums = sum_active_pes(fc_files)
attention_sums = sum_active_pes(attention_files)

# 
for file, total in fc_sums.items():
    print(f"{file}: {total}")
for file, total in attention_sums.items():
    print(f"{file}: {total}")
```
