```python
import csv
import re
import matplotlib.pyplot as plt

def convert_table_to_csv(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as fin:
        lines = fin.readlines()

    rows = []
    for line in lines:
        line = line.strip()
        if line.startswith('|') and line.endswith('|'):
            if re.match(r'^\|[-+\s]+$', line):
                continue
            cells = [cell.strip() for cell in line.split('|')[1:-1]]
            if any(cells):
                rows.append(cells)

    with open(output_file, 'w', encoding='utf-8', newline='') as fout:
        writer = csv.writer(fout)
        for row in rows:
            writer.writerow(row)

convert_table_to_csv('/content/compile_report_20.txt', 'compile_report_20.csv')
convert_table_to_csv('/content/compile_report_19.txt', 'compile_report_19.csv')
convert_table_to_csv('/content/compile_report_18.txt', 'compile_report_18.csv')
convert_table_to_csv('/content/compile_report_17.txt', 'compile_report_17.csv')
convert_table_to_csv('/content/compile_report_16.txt', 'compile_report_15.csv')
convert_table_to_csv('/content/compile_report_15.txt', 'compile_report_16.csv')
convert_table_to_csv('/content/compile_report_14.txt', 'compile_report_14.csv')
convert_table_to_csv('/content/compile_report_13.txt', 'compile_report_13.csv')
convert_table_to_csv('/content/compile_report_11.txt', 'compile_report_11.csv')
convert_table_to_csv('/content/compile_report_7.txt', 'compile_report_7.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

experiment = [7,11,13,14, 15, 16, 17, 18, 19, 20]

def load_data(metric_type):

    data = {}
    for ch in experiment:
        filename = f'filtered_compile_report_{ch}_{metric_type}.csv'
        df = pd.read_csv(filename)
        df['Samples/s'] = pd.to_numeric(df['Samples/s'], errors='coerce')
        df['Active PEs'] = pd.to_numeric(df['Active PEs'], errors='coerce')
        df['Compute Utilization'] = df['Compute Utilization'].str.rstrip('%')
        df['Compute Utilization'] = pd.to_numeric(df['Compute Utilization'], errors='coerce')
        data[ch] = df
    return data

data_fc = load_data('fc')
data_attention = load_data('attention')

def collect_metric(data, column):

    return [data[ch][column].dropna() for ch in experiment]

samples_fc = collect_metric(data_fc, 'Samples/s')
active_pe_fc = collect_metric(data_fc, 'Active PEs')
util_fc = collect_metric(data_fc, 'Compute Utilization')

samples_attention = collect_metric(data_attention, 'Samples/s')
active_pe_attention = collect_metric(data_attention, 'Active PEs')
util_attention = collect_metric(data_attention, 'Compute Utilization')


plt.figure(figsize=(10,6))
plt.boxplot(samples_fc, labels=experiment)
plt.title('FC Samples/s Across experiment')
plt.xlabel('experiment')
plt.ylabel('Samples/s')
plt.tight_layout()
plt.savefig('fc_samples_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(active_pe_fc, labels=experiment)
plt.title('FC Active PEs Across experiment')
plt.xlabel('experiment')
plt.ylabel('Active PEs')

plt.ylim(0, 2500)

plt.tight_layout()
plt.savefig('fc_active_pe_boxplot.png')
plt.show()


plt.figure(figsize=(10,6))
plt.boxplot(util_fc, labels=experiment)
plt.title('FC Compute Utilization (%) Across experiment')
plt.xlabel('experiment')
plt.ylabel('Compute Utilization (%)')
plt.tight_layout()
plt.savefig('fc_util_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(samples_attention, labels=experiment)
plt.title('Attention Samples/s Across experiment')
plt.xlabel('experiment')
plt.ylabel('Samples/s')
plt.tight_layout()
plt.savefig('attention_samples_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(active_pe_attention, labels=experiment)
plt.title('Attention Active PEs Across experiment')
plt.xlabel('experiment')
plt.ylabel('Active PEs')
plt.tight_layout()
plt.savefig('attention_active_pe_boxplot.png')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(util_attention, labels=experiment)
plt.title('Attention Compute Utilization (%) Across experiment')
plt.xlabel('experiment')
plt.ylabel('Compute Utilization (%)')
plt.tight_layout()
plt.savefig('attention_util_boxplot.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

reports = [7, 11, 13, 14, 15, 16, 17, 18, 19, 20]

sum_fc_list = []
sum_attention_list = []

for rep in reports:
    fc_file = f'filtered_compile_report_{rep}_fc.csv'
    attention_file = f'filtered_compile_report_{rep}_attention.csv'

    df_fc = pd.read_csv(fc_file)
    df_fc['Active PEs'] = pd.to_numeric(df_fc['Active PEs'], errors='coerce')
    sum_fc = df_fc['Active PEs'].sum()

    df_attention = pd.read_csv(attention_file)
    df_attention['Active PEs'] = pd.to_numeric(df_attention['Active PEs'], errors='coerce')
    sum_attention = df_attention['Active PEs'].sum()

    sum_fc_list.append(sum_fc)
    sum_attention_list.append(sum_attention)

for rep, s_fc, s_att in zip(reports, sum_fc_list, sum_attention_list):
    print(f"Report {rep} -> FC: {s_fc}, Attention: {s_att}, Sum: {s_fc + s_att}")

x = np.arange(len(reports))    #
width = 0.35                   #

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width/2, sum_fc_list, width, label='FC')
rects2 = ax.bar(x + width/2, sum_attention_list, width, label='Attention')

#
ax.set_xlabel('Report')
ax.set_ylabel('Active PEs')
ax.set_title('Report FC and Attentions Active PEs sum')
ax.set_xticks(x)
ax.set_xticklabels(reports)
ax.legend()

plt.tight_layout()
plt.savefig('active_pe_total_by_report.png')
plt.show()
```
